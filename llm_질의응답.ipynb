```python
def encode_questions_minilm(train):

    model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
    texts = []
    for _, row in train.iterrows():
        text = f"{row.질문_1}\n{row.질문_2}\n{row.답변_1}\n{row.답변_2}\n{row.답변_3}\n{row.답변_4}\n{row.답변_5}"
        texts.append(text)
    with torch.autocast(device_type='cuda'):
        embs = model.encode(texts, device=f'cuda', show_progress_bar=True).astype(np.float16)
    return embs
```

```python
encode_questions_minilm(train)
```
